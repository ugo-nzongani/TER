```python
import numpy as np
import time
from math import *
from qiskit import *
from qiskit.circuit import *
from qiskit.extensions import *
from qiskit.circuit.library import *
from qiskit.extensions.simulator.snapshot import snapshot
from qiskit.visualization import plot_histogram
from qiskit.quantum_info.operators import Operator
from qiskit.extensions.simulator.snapshot import snapshot
from scipy import optimize
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# Code pour la marche

'''
Renvoi la matrice de déplacement S
    - n : nombre de noeuds (nombre de positions possibles)
'''
def get_s(n):
    s0 = np.zeros((n,n))
    s1 = np.zeros((n,n))
    for i in range(n):
        s0[(i-1) % n][i] = 1
        s1[(i+1) % n][i] = 1
    s = np.kron(s0, [[1,0],[0,0]]) + np.kron(s1, [[0,0],[0,1]])
    return s

'''
Renvoi le coin operator U
    - n : nombre de noeuds (nombre de positions possibles)
    - coin : le coin operator qu'on souhaite utiliser
'''
def get_u(n, coin):
    return np.kron(np.identity(n), coin)

'''
Fait la marche quantique 
    - psi : vecteur d'état de la marche
     - n : nombre de noeuds (nombre de positions possibles)
     - coin : le coin operator qu'on souhaite utiliser
     - n_step : nombre de pas qu'on souhaite effectuer
'''
def quantum_walk(psi, n, coin, n_step, multiple_coin = False, coin_list = []):
    s = get_s(n)
    if(multiple_coin):
        u = get_u_multiple_coin(n, coin_list)
    else:
        u = get_u(n, coin)
    q = np.matmul(s,u)
    f = np.linalg.matrix_power(q, n_step)
    return np.matmul(f, psi)

'''
Décode le résultat de la marche
    - psi : vecteur d'état de la marche
    - n : nombre de noeuds (nombre de positions possibles)
'''
def decode(psi, n):
    distrib = {}
    for i in range(n):
        if(psi[2*i][0] != 0):
            distrib[i] = np.abs(psi[2*i][0])**2
        if(psi[2*i + 1][0] != 0):
            if i in distrib.keys():
                distrib[i] += np.abs(psi[2*i +1][0])**2
            else:
                distrib[i] = np.abs(psi[2*i +1][0])**2
    return distrib

'''
Normalise le vecteur psi
    - psi : vecteur d'état de la marche
'''
def normalize(psi):
    norme = 0
    for i in psi:
        norme += np.abs(i[0])**2
    if(norme != 0):
        psi = psi/np.sqrt(norme)
    return psi
    
'''
Initialise le vecteur psi
    - pos_init : tableau contenant les positions initiales
    - coin_init : tableau dont les valeurs sont les états initiaux de la pièce, [0,1] correspond à l'état (|0>+|1>)/sqrt(2)
'''
def init(pos_init, coin_init):
    psi = np.zeros((2*n, 1))
    for i in pos_init:
        for j in coin_init:
            if(j == -1):
                 psi[(2*i)+j*-1][0] = -1
            else:
                 psi[(2*i)+j][0] = 1
    return normalize(psi)

'''
Plot les positions en abscisses et les probas en ordonnées
    - n : nombre de noeuds (nombre de positions possibles)
    - distrib : distribution de probabilité
    - figsize : dimensions du plot
    - width : largeur des bars
'''
def plot_distrib(n, distrib, figsize = (6,6), width = 0.5):
    x = np.arange(n)
    y = []
    for i in range(n):
        if i in distrib.keys():
            y.append(distrib[i])
        else:
            y.append(float('nan'))
    plt.figure(figsize=figsize, dpi=80)
    plt.xlabel("Position")
    plt.ylabel("Probabilities")
    plt.bar(x, y, width = width, color='blue')

'''
Renvoi la matrice U
    n : nombre de noeuds
    coin_list : liste des coins operators, l'operateur d'indice i s'applique à la position i
'''
def get_u_multiple_coin(n, coin_list):
    coin_operator = np.zeros((2*n, 2*n))
    for i in range(n):
        a = np.zeros((n, n))
        a[i][i] = 1
        coin_operator += np.kron(a,coin_list[i])
    return coin_operator

# Code pour le circuit

'''
Renvoi les portes Z+ et Z-
    - n : nombre de qubits (log_2(nombre de position))
'''
def get_z_gates(n):
    pos = QuantumRegister(n, name="pos")
    # registre pour le coin operator
    coin = QuantumRegister(1, name="coin")
    aux = QuantumCircuit(pos, coin)
    '''
    for i in range(n):
        if(i == n-1):
            aux.x(q[i])
        else:
            c = [j for j in range(i+1,n)]
            aux.mcx(c, i, mode="noancilla")
    z_up = aux.to_gate(label="Z+")
    print("%s secondes pour Z+" % (time.time() - start_time))
    start_time = time.time()
    z_down = aux.inverse().to_gate(label="Z-")
    '''
    for i in range(n+1):
        if(i == n):
            aux.x(coin[0])
        else:
            ctrl = [coin[0]]
            for j in range(i,n):
                if(i != j):
                    ctrl.append(pos[j])
            aux.mct(ctrl, [pos[i]])

    z_up = aux.to_gate(label="Z+")
    z_down = aux.inverse().to_gate(label="Z-")
    return z_up, z_down

'''
Ajoute un pas au circuit
    - qc : registre quantique
    - pos : regisre de la position
    - coin : registre du coin
    - n : nombre de qubits (log_2(nombre de position))
    - z_up : porte Z+
    - z_down : porte Z-
'''
def step(qc, pos, coin, n, z_up, z_down, index, multiple_coin=False, coin_list=[]):
    

    if(multiple_coin):
        qubits = [pos[i] for i in range(n)]
        for i in range(len(coin)):
            qubits.append(coin[i])
        gate = get_variante_coin_gate(n, coin_list)
        qc.append(gate, qubits)
    else:
        
        qc.h(coin[0])
    qc.barrier()
    # Z+
    
    #t1 = [coin[0]]
    #t2 = [pos[i] for i in range(n)]
    #t3 = t1 + t2
    t1 = [pos[i] for i in range(n)]
    t1.append(coin[0])
    
    #qc.append(z_up.control(), t3)
    qc.append(z_up, t1)
    # NOT GATE
    qc.x(coin[0])
    # Z-
    #qc.append(z_down.control(), t3)
    qc.append(z_down, t1)
    # NOT GATE
    qc.x(coin[0])
    

'''
Renvoi le circuit qui implémente la marche
    - n : nombre de qubits (log_2(nombre de position))
    - n_step : nombre de pas
    - pos_init : entier correspondant à la position initiale
    - coin_init : tableau dont les valeurs sont les états initiaux de la pièce, [0,1] correspond à l'état (|0>+|1>)/sqrt(2)
'''
def circuit(n, n_step, pos_init, coin_init, multiple_coin=False, coin_list=[]):
    # registre pour la position
    pos = QuantumRegister(n, name="pos")
    # registre pour le coin operator
    coin = QuantumRegister(1, name="coin")
    # registre pour la mesure
    c = ClassicalRegister(n, name="y")

    qc = QuantumCircuit(pos, coin, c)
    # construction des portes Z+ et Z-
    z_up, z_down = get_z_gates(n)
    
    # initialisation du coin
    if(1 in coin_init):
        if(0 in coin_init):
            qc.h(coin[0])
        else:
            qc.x(coin[0])
    elif(-1 in coin_init):
        if(0 in coin_init):
            qc.x(coin[0])
            qc.h(coin[0])
        else:
            qc.x(coin[0])
            qc.z(coin[0])
    # initialisation de la position
    binary = bin(pos_init)[2:].zfill(n)
    # print(binary)
    #qc.h(pos[0])
    #qc.x(pos[1])
    for i in range(n):
        if(int(binary[i]) == 1):
            qc.x(pos[i])    
    qc.barrier()
    
    # on ajoute les pas
    
    
    for i in range(n_step):
        step(qc, pos, coin, n, z_up, z_down, i, multiple_coin, coin_list)
        
        qc.barrier()
    # on mesure seulement le registre de la position
    qc.measure(pos,c)
    return qc


'''
Renvoi la porte faisant varier le coin operator en fonction de la position
    coin_list : liste des différents coin operator
    nombre de qubits (log_2(nombre de position))
'''
def get_variante_coin_gate(n, coin_list):
    # registre pour la position
    pos = QuantumRegister(n, name="pos")
    # registre pour le coin operator
    coin = QuantumRegister(1, name="coin") # A CHANGER SI LE COIN AGIT SUR PLUS DE 1 QUBIT
    qc = QuantumCircuit(pos, coin)
    
    # les qubits sur lesquels le coin operator agit
    gate_qubits = [coin[k] for k in range(len(coin))]
    ctrl_qubits = [pos[k] for k in range(n)]
    qubits = ctrl_qubits + gate_qubits
    next_x = []
    for i in range(int(pow(2,n))):
        binary = bin(i)[2:].zfill(n)
        for j in range(n):
            if(i == 0):
                qc.x(pos[j])
                next_x.append(1)
            elif next_x[n-j-1] == int(pow(2,j)):
                qc.x(pos[n-j-1])
                next_x[n-j-1] = 1
            else:
                next_x[n-j-1] += 1
        gate = UnitaryGate(coin_list[int(binary,2)], label=binary)
        qc.append(gate.control(n), qubits)

    
    return qc.to_gate(label="C")

'''
Renvoi la distribution de probabilité de la mesure du circuit lancé n_shots fois
    - circ : circuit
    - n_shots : nombre de fois qu'on lance le circuit
'''
def get_results(circ, n_shots):
    # résultats des mesures du circuit
    '''
    backend = BasicAer.get_backend('qasm_simulator')
    #job = execute(circ, backend, shots=n_shots)
    #counts = dict(job.result().get_counts(circ))
    
    job = execute(circ, backend=backend, shots=n_shots)

    counts = dict(job.result().get_counts(circ))
    '''

    backend_sim = Aer.get_backend('qasm_simulator')

    job_sim = backend_sim.run(transpile(circ, backend_sim), shots=n_shots)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(circ)
    # pour remettre les indices dans le bon ordre
    #measure = {}
    distrib = {}
    for i in counts.keys():
        #measure[int(i[::-1],2)] = counts[i]
        distrib[int(i[::-1],2)] = counts[i]/n_shots
    return distrib

'''
Compare les distributions de probabilité obtenus avec le marche et le circuit
    - distrib_qw : distribution de probabilité obtenue avec la marche
    - distrib_circ : distribution de probabilité obtenue avec le circuit
'''
def compare_distrib(distrib_qw, distrib_circ):
    res = {}
    for i in distrib_qw:
        if(i in distrib_circ.keys()):
            res[i] = np.abs(distrib_qw[i] - distrib_circ[i])
        else:
            res[i] = distrib_qw[i]
    keys = distrib_qw.keys()
    for i in distrib_circ:
        if(i not in keys):
            if(i in distrib_qw.keys()):
                res[i] = np.abs(distrib_qw[i] - distrib_circ[i])
            else:
                res[i] = distrib_circ[i]
    return res
```

## Variation du coin operator en fonction de la position

```python
n = 2 # nombre de noeuds (puissance de 2 obligatoire pour que le circuit fonctionne)
n_step = 1 # nombre de steps

coin_list = [] # liste des coins operators qu'on va utiliser
for i in range(n):
    if(i%2 == 0): # Hadamard pour les positions paires
        coin_list.append(1/np.sqrt(2) * np.array([[1,1], [1, -1]]))
    else: # Z pour les positions impaires
        coin_list.append(np.array([[1,0], [0, -1]]))

pos_init = [0] # on commence au noeud 0
coin_init = [0]

psi = init(pos_init, coin_init)
p = quantum_walk(psi, n, coin, n_step, multiple_coin=True, coin_list=coin_list)
distrib_qw = decode(p,n)

# circuit, ne fonctionne pas lorsque la position initiale est une superposition à cause des états
# intriqués (il faut faire l'initialisation à la main dans ce cas)
circ = circuit(ceil(log(n,2)), n_step, pos_init[0], coin_init, multiple_coin=True, coin_list=coin_list)
distrib_circ = get_results(circ, 1024)
circ.decompose().draw()
#circ.draw("mpl")
```
