```python
# code pour la marche

def get_s(n):
    s0 = np.zeros((n,n))
    s1 = np.zeros((n,n))
    for i in range(n):
        s0[(i-1) % n][i] = 1
        s1[(i+1) % n][i] = 1
    s = np.kron(s0, [[1,0],[0,0]]) + np.kron(s1, [[0,0],[0,1]])
    return s

def get_u(n, coin):
    return np.kron(np.identity(n), coin)

def quantum_walk(psi, n, coin, n_step):
    s = get_s(n)
    u = get_u(n, coin)
    q = np.matmul(s,u)
    f = np.linalg.matrix_power(q, n_step)
    return np.matmul(f, psi)

def decode(psi, n):
    res = {}
    for i in range(n):
        if(psi[2*i][0] != 0):
            res[i] = np.abs(psi[2*i][0])**2
        if(psi[2*i + 1][0] != 0):
            if i in res.keys():
                res[i] += np.abs(psi[2*i +1][0])**2
            else:
                res[i] = np.abs(psi[2*i +1][0])**2
    return res
```

```python
# code pour le circuit

def get_z_gates(n):
    q = QuantumRegister(n, name="x")
    aux = QuantumCircuit(q)
    for i in range(n):
        if(i == n-1):
            aux.x(q[i])
        else:
            c = [j for j in range(i+1,n)]
            aux.mcx(c, i, mode="noancilla")
    z_up = aux.to_gate(label="Z+")
    z_down = aux.inverse().to_gate(label="Z-")
    return z_up, z_down

def step(qc, pos, coin, n, z_up, z_down):
    qc.h(coin[0])
    # Z+
    t1 = [coin[0]]
    t2 = [pos[i] for i in range(n)]
    t3 = t1 + t2
    qc.append(z_up.control(), t3)
    # NOT GATE
    qc.x(coin[0])
    # Z-
    qc.append(z_down.control(), t3)
    # NOT GATE
    qc.x(coin[0])
    
def circuit(n, n_step):
    # registre pour la position
    pos = QuantumRegister(n, name="pos")
    # registre pour le coin operator
    coin = QuantumRegister(1, name="coin")
    # registre pour la mesure
    c = ClassicalRegister(n, name="y")

    qc = QuantumCircuit(pos, coin, c)
    # construction des portes Z+ et Z-
    z_up, z_down = get_z_gates(n)

    for i in range(n_step):
        step(qc, pos, coin, n, z_up, z_down)
        
    # on mesure seulement le registre de la position
    qc.measure(pos,c)
    return qc
```

```python
n = 4 # nombre de noeuds
n_step = 1 # nombre de step
```

```python
# marche
coin = (1/np.sqrt(2)) * np.array([[1,1], [1, -1]]) # Hadamard coin
psi = np.zeros((2*n, 1))
psi[0][0] = 1 # initialisation de psi

p = quantum_walk(psi, n, coin, n_step)
res_qw = decode(p,n)
```

```python
# circuit
circ = circuit(ceil(log(n,2)), n_step)
circ.draw("mpl")
```

```python
# r√©sultats des mesures du circuit
backend = BasicAer.get_backend('qasm_simulator')
job = execute(circ, backend, shots=1024)
counts = dict(job.result().get_counts(circ))
# pour remettre les indices dans le bon ordre
d = {}
for i in counts.keys():
    d[int(i[::-1],2)] = counts[i]
```

```python
plot_histogram(d)
```

```python
# pour le plot de la marche
x = np.arange(n)
y = []
for i in range(n):
    if i in res_qw.keys():
        y.append(res_qw[i])
    else:
        y.append(0)
```

```python
plt.figure(figsize=(9, 9), dpi=80)
plt.plot(x,y,color='orange',linewidth=1)
plt.xlabel("Position")
plt.ylabel("Probabilities")
```
